```python
# import modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# assign data
dataFrame = pd.DataFrame({
    'Name': [' RACHEL  ', ' MONICA  ', ' PHOEBE  ', '  ROSS    ', 'CHANDLER', ' JOEY    ', ' DANNY ', ' DAVID '],
    'Age': [30, 35, 37, 33, 34, 30, 28, 38],
    'Salary': [100000, 93000, 88000, 120000, 94000, 95000, 105000, 98000],
    'JOB': ['DESIGNER', 'CHEF', 'MASUS', 'PALENTOLOGY', 'IT', 'ARTIST', 'DEVELOPER', 'DOCTOR']
})

# remove leading/trailing spaces in Name and JOB
dataFrame['Name'] = dataFrame['Name'].str.strip()
dataFrame['JOB'] = dataFrame['JOB'].str.strip()

# filter dataframe
filtered_df = dataFrame.loc[
    (dataFrame['Salary'] >= 100000) &
    (dataFrame['Age'] < 40) &
    (dataFrame['JOB'].str.startswith('D')),
    ['Name', 'Salary', 'JOB']
]

# plot the graph
plt.figure(figsize=(8, 5))
sns.barplot(data=filtered_df, x='Name', y='Salary', hue='JOB')
plt.title('High Salary Employees with JOB Starting with "D"')
plt.xlabel('Employee Name')
plt.ylabel('Salary')
plt.tight_layout()
plt.show()

```
